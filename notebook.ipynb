## Task 2

Create a visualization that shows how many products are repeat purchases. Use the visualization to:
- State which category of the variable repeat purchases has the most observations.
- Explain whether the observations are balanced across categories of the variable repeat purchases.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
pet_supplies_2212 = pd.read_csv('pet_supplies_2212.csv')

# Get the frequency of repeat purchases
repeat_counts = pet_supplies_2212['repeat_purchase'].value_counts()

# Plot a bar chart
repeat_counts.plot(kind='bar', color = 'red')

# Add labels and title
plt.xlabel('Repeat Purchase')
plt.ylabel('Frequency')
plt.title('Number of Products by Repeat Purchase')

# Show the plot
plt.show()
```

From the plot above, you can see that there are more products that are repeat purchases (1) than products that are repeat purchases (0). This suggests that most customers buy thesame products each time they shop for their pets. The observations are not balanced across categories of the variable repeat purchases.

## Task 3

- Describe the distribution of all of the sales. Your answer must include a visualization.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
pet_supplies_2212 = pd.read_csv('pet_supplies_2212.csv')

# Plot a histogram of sales
pet_supplies_2212['sales'].hist(bins=20, color = 'red')

# Add labels and title
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')

#Removing gridlines
plt.grid(b=False)

# Show the plot
plt.show()
```

From the histogram above, you can see that the distribution of sales is right-skewed, meaning that most sales are on the lower end of the scale, with a few outliers on the higher end. The mean (or average) of sales is **996.60** but it is not a good measure of the center of the distribution, because it is influenced by the outliers. A better measure of the center is the median (or middle value), which is **1000.83**. The standard deviation (or measure of spread) of sales is **323.95**, which means that there is a lot of variation in sales among different products.

## Task 4

- Describe the relationship between repeat purchases and sales. Your answer must include a visualization to demonstrate the relationship.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
pet_supplies_2212 = pd.read_csv('pet_supplies_2212.csv')

# Plot a scatter plot of repeat_purchase vs sales
pet_supplies_2212.plot(kind='scatter', x='repeat_purchase', y='sales', color = 'red')

# Add labels and title
plt.xlabel('Repeat Purchase')
plt.ylabel('Sales')
plt.title('Relationship between Repeat Purchase and Sales')

# Show the plot
plt.show()
```

- From the scatter plot above, it is evident that there is no clear relationship between repeat_purchase and sales. The points are scattered randomly across the plane, with no discernible pattern or trend. This means that repeat_purchase does not affect sales, or vice versa. 
- The correlation coefficient (or measure of strength and direction of the relationship) between repeat_purchase and sales is -0.01, which is very close to zero. This indicates that there is no linear relationship between the two variables.
